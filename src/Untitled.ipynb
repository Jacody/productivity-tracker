```python
import tkinter as tk
from tkinter import messagebox
import json

# Datei zur Speicherung der To-Do-Liste
TODO_FILE = "todo_list.json"

def load_tasks():
    try:
        with open(TODO_FILE, "r") as file:
            return json.load(file)
    except (FileNotFoundError, json.JSONDecodeError):
        return []

def save_tasks():
    with open(TODO_FILE, "w") as file:
        json.dump(tasks, file, indent=4)

def add_task():
    task = task_entry.get()
    if task:
        tasks.append({"task": task, "done": False, "subtasks": []})
        update_list()
        save_tasks()
        task_entry.delete(0, tk.END)
    else:
        messagebox.showwarning("Fehler", "Bitte eine Aufgabe eingeben!")

def add_subtask():
    try:
        selected_index = task_list.curselection()[0]
        subtask = subtask_entry.get()
        if subtask:
            tasks[selected_index]["subtasks"].append({"task": subtask, "done": False})
            update_list()
            save_tasks()
            subtask_entry.delete(0, tk.END)
        else:
            messagebox.showwarning("Fehler", "Bitte eine Teilaufgabe eingeben!")
    except IndexError:
        messagebox.showwarning("Fehler", "Bitte eine Hauptaufgabe auswählen!")

def update_list():
    task_list.delete(0, tk.END)
    for i, task in enumerate(tasks):
        status = "✔ " if task["done"] else "✖ "
        task_list.insert(tk.END, f"{status} {task['task']}")
        for subtask in task["subtasks"]:
            sub_status = "✔ " if subtask["done"] else "✖ "
            task_list.insert(tk.END, f"    {sub_status} {subtask['task']}")

def toggle_task():
    try:
        selected_index = task_list.curselection()[0]
        tasks[selected_index]["done"] = not tasks[selected_index]["done"]
        update_list()
        save_tasks()
    except IndexError:
        messagebox.showwarning("Fehler", "Bitte eine Aufgabe auswählen!")

def delete_task():
    try:
        selected_index = task_list.curselection()[0]
        del tasks[selected_index]
        update_list()
        save_tasks()
    except IndexError:
        messagebox.showwarning("Fehler", "Bitte eine Aufgabe auswählen!")

# GUI erstellen
root = tk.Tk()
root.title("To-Do-Liste")
root.geometry("400x500")

task_entry = tk.Entry(root, width=40)
task_entry.pack(pady=10)

task_add_button = tk.Button(root, text="Hinzufügen", command=add_task)
task_add_button.pack()

task_list = tk.Listbox(root, width=50, height=15)
task_list.pack(pady=10)

subtask_entry = tk.Entry(root, width=40)
subtask_entry.pack(pady=5)

subtask_add_button = tk.Button(root, text="Teilaufgabe hinzufügen", command=add_subtask)
subtask_add_button.pack()

button_frame = tk.Frame(root)
button_frame.pack()

task_done_button = tk.Button(button_frame, text="Erledigt", command=toggle_task)
task_done_button.grid(row=0, column=0, padx=5)

task_delete_button = tk.Button(button_frame, text="Löschen", command=delete_task)
task_delete_button.grid(row=0, column=1, padx=5)

# Daten laden
tasks = load_tasks()
update_list()

root.mainloop()
```
