```python
import os
import sys

# TensorFlow- und Mediapipe-Warnungen unterdrücken
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'  # TensorFlow nur kritische Fehler anzeigen
import absl.logging
absl.logging.set_verbosity(absl.logging.ERROR)  # Mediapipe Logging deaktivieren

# OpenCV-Warnungen unterdrücken (Alternative Methode)
os.environ["OPENCV_LOG_LEVEL"] = "ERROR"
import cv2
try:
    cv2.setLogLevel(3)  # Falls vorhanden, Logging-Level setzen
except AttributeError:
    pass  # Falls nicht verfügbar, einfach ignorieren

# 🔥 WICHTIG: STDERR komplett unterdrücken (ALLE Fehler in eine Datei umleiten)
sys.stderr = open(os.devnull, 'w')  # Alle Fehler ausblenden (alternativ: 'error_log.txt')


import sys
from datetime import datetime, timedelta
from PyQt5.QtWidgets import QApplication, QWidget, QVBoxLayout, QPushButton, QLabel
from PyQt5.QtCore import QTimer
from face_detection import FaceDetectorApp

class MainApp(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Face Detection Launcher")
        self.setGeometry(200, 200, 400, 300)

        # Layout erstellen
        layout = QVBoxLayout()

        # "Start"-Button
        self.start_button = QPushButton("Start Face Detection", self)
        self.start_button.clicked.connect(self.start_face_detection)

        # "Hold"-Button
        self.hold_button = QPushButton("Hold", self)
        self.hold_button.setStyleSheet("background-color: none;")  # Standardfarbe
        self.hold_button.clicked.connect(self.toggle_hold_mode)  # Button-Logik

        # Label für Status-Anzeige
        self.status_label = QLabel("Status: Nicht gestartet", self)

        # Timer-Label
        self.timer_label = QLabel("Timer: 00:00:00", self)

        layout.addWidget(self.start_button)
        layout.addWidget(self.hold_button)
        layout.addWidget(self.status_label)
        layout.addWidget(self.timer_label)
        self.setLayout(layout)

        self.face_detector_window = None  # Wird beim Start zugewiesen
        self.start_time = None  # Startzeit des Face Detection Fensters
        self.mode = 1  # Hold-Status (1 = normal, 0 = Hold aktiviert)
        self.status = 0  # Gesichtserkennungs-Status (1 = erkannt, 0 = nicht erkannt)

        # Timer zur Aktualisierung der GUI-Zeit
        self.active_time = timedelta(seconds=0)  # Gesamtzeit
        self.last_timer_start = None  # Zeitpunkt des letzten Starts
        self.gui_timer = QTimer()
        self.gui_timer.timeout.connect(self.update_timer_label)
        self.gui_timer.start(1000)  # Timer läuft immer, zählt aber nicht immer

    def start_face_detection(self):
        """Startet die Gesichtserkennung und speichert die Startzeit."""
        if not self.face_detector_window:
            # Speichere die Startzeit (Datum + Uhrzeit)
            self.start_time = datetime.now()
            start_time_str = self.start_time.strftime("%Y-%m-%d %H:%M:%S")

            print(f"📅 Startzeit erfasst: {start_time_str}")  # Ausgabe ins Terminal
            print(f"Mode= {self.mode} Status= {self.status} Time= {self.active_time}")

            self.face_detector_window = FaceDetectorApp()
            self.face_detector_window.status_changed.connect(self.update_status)  # Signal verbinden
            self.face_detector_window.show()

    def toggle_hold_mode(self):
        """Schaltet den Hold-Modus an/aus"""
        if self.mode == 1:
            self.mode = 0
            self.hold_button.setStyleSheet("background-color: red; color: white;")
            #print("🛑 Hold aktiviert | Mode= 0")
            print(f"Mode= {self.mode} Status= {self.status} Time= {self.active_time}")

        else:
            self.mode = 1
            self.hold_button.setStyleSheet("background-color: none;")
            #print("✅ Hold deaktiviert | Mode= 1")
            print(f"Mode= {self.mode} Status= {self.status} Timer= {str(self.active_time).split('.')[0]} Time= {datetime.now().strftime('%H:%M:%S')}")


        self.update_timer_label()

    def update_status(self, status):
        """Aktualisiert den Status"""
        self.status = status
        status_text = "Gesicht erkannt" if status == 1 else "Kein Gesicht erkannt"
        self.status_label.setText(f"Status: {status_text}")
        #print(f"🔄 Status aktualisiert: {status}")
        print(f"Mode= {self.mode} Status= {self.status} Time= {self.active_time}")
        

        self.update_timer_label()

    def update_timer_label(self):
        """Aktualisiert die Timer-Anzeige in der GUI"""
        if self.status == 1 and self.mode == 1:  # Gesicht erkannt UND Hold deaktiviert
            if self.last_timer_start is None:  # Falls der Timer nicht läuft
                self.last_timer_start = datetime.now()
            elapsed_time = datetime.now() - self.last_timer_start
            total_time = self.active_time + elapsed_time
        else:
            if self.last_timer_start is not None:  # Falls Timer lief, Zeit speichern
                self.active_time += datetime.now() - self.last_timer_start
                self.last_timer_start = None  # Timer stoppen
            total_time = self.active_time  # Pausierte Zeit beibehalten

        # Zeit in HH:MM:SS anzeigen
        time_str = str(total_time).split(".")[0]  
        self.timer_label.setText(f"Timer: {time_str}")

# Hauptprogramm
if __name__ == "__main__":
    app = QApplication(sys.argv)
    main_window = MainApp()
    main_window.show()
    sys.exit(app.exec_())

```
